#Question1: How many Movies and How many Shows are there on Netflix? Plot a graph to show the difference

```python
movie = data["type"]=='MOVIE'
number_of_movies = sum(movie==True)
show = data['type'] == 'SHOW'
number_of_shows = sum(show==True)
```

After counting the number of each, we can now plot the graph

```python
import matplotlib.pyplot as plt
x = ["Movies","Shows"]
y = [number_of_movies, number_of_shows]
print("Number of Movies is: "+str(number_of_movies))
print("Number of Shows is: "+str(number_of_shows))
plt.bar(x,y)
plt.show()
```

#Question 2: In Which year most movies or series were released?

To calculate in which year the most movies or series were released, we are going to run the loop to go through the data and calculat the count of each year and then plot that in graph

```python
years = []
year_count = []
release_year = data["release_year"].drop_duplicates()


for year in release_year:
    count = 0
    for index in range(0,len(data["release_year"])):
        if year == data["release_year"][index]:
            count +=1
    years.append(year)
    year_count.append(count)
print(years)
```

```python
#Creating table for each year
x = years
y = year_count
# create a new dataframe
d = {"Year":years,"Release per year":year_count}
df = pd.DataFrame(d)
df.sort_values("Release per year",axis = 0,inplace = True)
df.tail()
```

From the above we can see that most released in 2019, then we can plot the graph

```python
x = years
y =year_count
fig,ax = plt.subplots()
ax.plot(x,y,linewidth = 3.0)
plt.show()
```
